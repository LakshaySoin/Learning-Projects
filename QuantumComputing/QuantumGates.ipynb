```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
qc = QuantumCircuit(3)
```

```python
qc.draw(output='mpl')
```

```python
qc.x(0)
qc.y(1)
qc.z(2)
qc.draw(output='mpl')
```

```python
from qiskit.visualization import plot_bloch_multivector
```

```python
simulator = Aer.get_backend('statevector_simulator')
res = execute(qc, backend=simulator).result()
```

```python
plot_bloch_multivector(res.get_statevector())
```

```python
qc.h([0, 1, 2])
```

```python
qc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
res = execute(qc, backend=simulator).result()
plot_bloch_multivector(res.get_statevector())
```

```python
qc2 = QuantumCircuit(3)
qc2.h([0, 1, 2])
qc2.x(0)
qc2.y(1)
qc2.z(2)
simulator = Aer.get_backend('statevector_simulator')
res = execute(qc2, backend=simulator).result()
plot_bloch_multivector(res.get_statevector())
```

```python
qc2.draw(output='mpl')
```
